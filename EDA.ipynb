```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import datetime as dt

%matplotlib inline
```

```python
df = pd.read_csv('./data/df_with_tags.gz')
```

```python
df['Review_Date'] = df['Review_Date'].apply(lambda x: dt.datetime.strptime(x,'%Y-%m-%d'))
```

```python
df.shape
```

```python
df.columns
```

```python
df.count()
```

```python
fig, ax = plt.subplots(figsize=(14,10))
df.hist(bins=25, ax=ax)
plt.tight_layout()
```

```python
df[['Reviewer_Score', 'Reviewer_Nationality'
    ]].groupby('Reviewer_Nationality').describe().sort_values(
        ('Reviewer_Score', 'count'), ascending=False).head(10)
```

```python
df[['Review_Total_Positive_Word_Counts', 'Reviewer_Nationality'
    ]].groupby('Reviewer_Nationality').describe().sort_values(
        ('Review_Total_Positive_Word_Counts', 'count'), ascending=False).head(10)
```

```python
df[['Review_Total_Negative_Word_Counts', 'Reviewer_Nationality'
    ]].groupby('Reviewer_Nationality').describe().sort_values(
        ('Review_Total_Negative_Word_Counts', 'count'), ascending=False).head(10)
```

```python
count_reviews_hotel = pd.DataFrame(
    df.groupby(['Hotel_Name', 'Country', 'Average_Score']).apply(lambda x: len(x)).reset_index())

count_reviews_hotel.rename(columns={0: "Reviews"}, inplace=True)

count_reviews_hotel[count_reviews_hotel.Reviews > 100].sort_values('Average_Score', ascending=False).head(10)
```

```python
avg_country = df[['Diff', 'Reviewer_Nationality'
                  ]].groupby('Reviewer_Nationality').describe().sort_values(
                      ('Diff', 'count'), ascending=False)

avg_country.head(10)
```

```python
# Most positive countries (at least 100 reviews)
avg_country[avg_country['Diff']['count'] > 100].sort_values(('Diff', 'mean'), ascending=False).head(10)
```

```python
# Most negative countries (at least 100 reviews)
avg_country[avg_country['Diff']['count'] > 100].sort_values(('Diff', 'mean'), ascending=True).head(10)
```

```python
fig, ax = plt.subplots(1,1, figsize=(14,5))
ax.scatter(df['Average_Score'], df['Review_Total_Negative_Word_Counts'], c = 'r', alpha = .2, label = 'Negative')
ax.scatter(df['Average_Score'], df['Review_Total_Positive_Word_Counts'], c = 'b', alpha = .2, label = 'Positive')
ax.legend()
```

### Tags Analysis

#### People with Pets

```python
# Barcelona the city with better hotels for people with pets.
pd.pivot_table(df[['Reviewer_Score', 'Pet', 'City']],
               values='Reviewer_Score', index=['Pet'], columns=['City'],
               aggfunc=np.mean)
```

#### Trip purpose

```python
df[['Reviewer_Score', 'Purpose']].groupby('Purpose').mean()
```

```python
# Barcelona and Vienna have the best hotels for Leisure and Business stays

pd.pivot_table(df[['Reviewer_Score', 'Purpose', 'City']],
               values='Reviewer_Score', index=['Purpose'], columns=['City'],
               aggfunc=np.mean)
```

```python
# Barcelona and Vienna have the best hotels for Leisure and Business stays

pd.pivot_table(df[['Diff', 'Purpose', 'City']],
               values='Diff', index=['Purpose'], columns=['City'],
               aggfunc=np.mean)
```

#### With whom

```python
# Barcelona and Vienna have the best hotels for Leisure and Business stays

pd.pivot_table(df[['Diff', 'Whom', 'City']],
               values='Diff', index=['Whom'], columns=['City'],
               aggfunc=np.mean)
```
